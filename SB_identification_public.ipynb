# Plan

1) Separate magnetic field data by month.
2) identify timegaps in magnetic field timeseries
3) In order to provide a meaningful comparison between intervals at different distances
   we should resample the timeseries at a constant same sampling time. I suggest we resample to 1 second.
4) We use the code to identify events using the resampled timeseries
5) Once we have identified events, we remove from our dataframe all those events that correspond to times 
   related to timegaps (Note that we will still have data for those times since we linearly interpolated, but we should remove those). 
   I suggest we impose a theshold of 10 seconds for timegaps i.e., we ony remove gaps of duration > 10 second.
6) We plot low resolution magnetic field data by month. We identify HCS crossing and remove all the events that were identified and correspond to HCS crossings or other transients.
7) We consider the duration $\delta \tau$ of the identified events. However, we need the scale of the events in units of the ion inertial length $d_i$. We convert temporal scales to spatial scales through the modified Taylor's hypothesis
   $\ell =V_{tot}* \delta \tau$, where $V_{tot} = |\boldsymbol{V_{sw} + \boldsymbol{V}_{a} - \boldsymbol{V}_{sc}}|$
   
8) We can not use PSP data beyond 0.5 au due to poor data quality. Between 0.5 and 1 au I suggest we use SoLO data.
